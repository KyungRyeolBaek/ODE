Solve the following initial value differential equation $$ y'=\lambda y,  \quad  y(0)=y_0$$

Fuction Define

```python
# analysis 함수 
def anal(x, lamb):
    return np.exp(lamb*x)

# n : w 개수, w : weight
def y(x, w, n = 4):
    y = 1         # y = 1 + w1*(x) + w2*(x**2) + w3*(x**3) + w4*(x**4)
    for i in range(1, n+1):
        y += w[i-1]*(x**i)
    
    return y

def yprime(x, w, n = 4):
    yprime = 0    # yprime = w1 + 2*w2*x + 3*w3*(x**2) + 4*w4*(x**3)
    for i in range(1, n + 1):
        yprime += i*w[i-1]*(x**(i-1))

    return yprime

def f(y, lamb):
    return lamb*y
```

Parameter

```python
# 최대 x 범위 : 0 <= x < x_max, lambda 값
x_max = 1
x = tf.Variable(np.array([[i/100] for i in range(0, x_max * 100, 1)], dtype = np.float32))
lamb = -2

# 차수, weight 개수
n = 4
w = tf.Variable(tf.random.normal(shape=(n,), dtype=tf.float32), trainable=True) # w = [weight1, weight2, weight3, weight4]

# 변화 w 값, loss 저장
w_trace = []
w_trace.append(w.numpy())
loss_trace = []

#  학습률 조정
learning_rate = 0.01
# EPOCH = 100

# 원하는 loss 값에 도달할 때 까지 Epoch 횟수 올리기
Epoch = 0
target_loss = 0.001
```

Train

```python
import tensorflow as tf
from tensorflow import keras

class expansion(keras.layers.Layer):
    def __init__(self, units=32, input_dim=32, lamb = 1):
        super(expansion, self).__init__()
        self.w = self.add_weight(
            shape=(input_dim, units), initializer="random_normal", trainable=True
        )
        self.b = self.add_weight(shape=(input_dim,), initializer="zeros", trainable=True)
    
    # analysis 함수 
    def anal(x, lamb):
        return np.exp(lamb*x)
    
    def f(y, lamb):
        return lamb*y

    # n : w 개수, w : weight
    def y(self, inputs):
        y = tf.matmul(self.w, inputs) + self.b
        return y

    def yprime(self, inputs):
        with tf.GradientTape() as tape_:
            z = f(y, lamb)
        yprime = tape_.gradient(z, [x])

        return yprime

    def call(self, inputs):
        # Instantiate an optimizer.
        optimizer = keras.optimizers.SGD(learning_rate=1e-3)
        epochs = 2
        for epoch in range(epochs):
            print("\nStart of epoch {}".format(epoch))
            for inputs in tf.Variable([float(x_val)/100 for x_val in range(100)], trainable=False):
                with tf.GradientTape() as tape:
                    predict = f(y(inputs), lamb)
                    loss = tf.math.sqrt(tf.square(predict - f(yprime(inputs))))
                grads = tape.gradient(loss, predict.trainable_weights)
                optimizer.apply_gradients(zip(grads, predict.trainable_weights))


        return predict.trainable_weights



# x = tf.ones((2, 2))
x_val = 0.1
units = 4
input_dim = 1
x = tf.Variable([[float(x_val)**(p+1)] for p in range(units)], trainable=False)
expansion_layer = expansion(units, 1)
print(x.shape)
y = tf.ones((4, 1))
print(y.shape)
expansion_layer(1)
# y = expansion_layer(x)
# print(y)
# assert linear_layer.weights == [linear_layer.w, linear_layer.b]
```

```python
x = tf.Variable([[x_val**(p+1)] for p in range(units)], trainable=False)
y = tf.Variable([x_val**(p+1) for p in range(units)], trainable=False)
x.shape, y.shape
print(tf.matmul(y, x))
```

```python
optimizer = keras.optimizers.SGD(learning_rate=1e-3)
# 학습 과정
while True:
    Epoch += 1
    for x_unit in x:
        with tf.GradientTape() as tape:
            # 예측 값
            predict = f(y(x_unit, w, n), lamb)
            loss = tf.math.sqrt(tf.square(predict - f(yprime(x_unit, w, n), lamb)))
            loss_trace.append(loss.numpy())
            gradients = tape.gradient(loss, w)
            # # 오차
            # w_trace.append(w.numpy())
            # w = tf.Variable(w - learning_rate*gradients)
        if loss < target_loss:
            break
        optimizer.apply_gradients(zip(gradients, predict.trainable_weights))
    if Epoch % 1000 == 0:
        # print('Epoch : {}, Loss : {:.4f}'.format(Epoch, loss))
        print(Epoch, loss)
```
